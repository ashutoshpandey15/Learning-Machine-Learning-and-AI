# Array Operations

#### ARITHEMETIC OPERATIONS

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
print(a1)
print(a2)
```

```python
# Size of the arrays must be the same 
a3 = a1+a2
a3
```

#### Broadcasting 

```python
l = [10,20,30,40]
arr = np.array(l)
arr
```

```python
# Suppose we have to add 5 to each element 
arr + 5
# This is called broadcasting
```

### Matrix Operations

```python
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])
```

```python
A
```

```python
B
```

```python
# This is invalid in matrix multiplication
A*B
```

```python
# This gives accurate matrix multiplication
A @ B
```

```python
#  Transpose  = Rows become columns and columns becomes row
A.T
```

```python
B.T
```

```python
(A @ B).T
```
