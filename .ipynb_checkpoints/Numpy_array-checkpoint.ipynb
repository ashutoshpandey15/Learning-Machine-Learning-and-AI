```python
import numpy as np
```

## Basic Array Creation

```python
# Creating a simple numpy arrays
arr = np.array([1,2,3,4])
arr
```

```python
# Creating a simple list in python
list1 = [1,2,3,4, "hi", 5.9]
list1
```

```python
# The numpy array converts all data into simple datatype
np.array(list1)
```

```python
# Trying to create 3 dimensional lists is not possible
l = [ [1,2,3], [4,5,6], [7,8,9] ]
l
```

```python
# To convert the above list "l" into matrix, we can use array function
arr = np.array(l)
arr
```

# Array Generation Function

```python
# Creating an array using range function
arr1 = np.arange(1,11)
arr1
```

```python
# with steps
arr2 = np.arange(1,11,2)
arr2
```

```python
# Creating array of zeros
arr = np.zeros((4,8))
arr
```

```python
# Creating array of ones
arr = np.ones((4,8))
arr
```

```python
# Creating arrays with linspace
arr4 = np.linspace(1,5,2)        
print(arr4)
# That means that we need 2 elements between 1 and 5 with even space or gap (Linear space).

arr5 = np.linspace(1,5,3)
print(arr5)
```
